```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# Set up the path to your data directory
data_dir = '/content/drive/MyDrive/Veg_dat/Veg_test'
```

```python
# Image data generator for loading and augmenting data
train_datagen = ImageDataGenerator(rescale=1.0/255, validation_split=0.2)  # Normalizing pixel values

# Load training data
training_set = train_datagen.flow_from_directory(
    data_dir,
    target_size=(64, 64),  # Resize images
    batch_size=32,
    class_mode='categorical',  # For multi-class classification
    subset='training',  # Set as training data
    seed=42
)

# Load validation data
validation_set = train_datagen.flow_from_directory(
    data_dir,
    target_size=(64, 64),  # Resize images
    batch_size=32,
    class_mode='categorical',  # For multi-class classification
    subset='validation',  # Set as validation data
    seed=42
)

# Create the CNN model
cnn = Sequential()
cnn.add(Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)))
cnn.add(MaxPooling2D(pool_size=(2, 2)))
cnn.add(Conv2D(64, (3, 3), activation='relu'))
cnn.add(MaxPooling2D(pool_size=(2, 2)))
cnn.add(Flatten())
cnn.add(Dense(128, activation='relu'))
cnn.add(Dense(len(training_set.class_indices), activation='softmax'))  # Output layer

# Compile the model
cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
training_history = cnn.fit(
    x=training_set,
    validation_data=validation_set,
    epochs=32
)

# Evaluate the model
test_loss, test_acc = cnn.evaluate(validation_set)
print('Validation accuracy:', test_acc)

# Optionally, plot the training history
plt.plot(training_history.history['accuracy'], label='Training Accuracy')
plt.plot(training_history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.show()

plt.plot(training_history.history['loss'], label='Training Loss')
plt.plot(training_history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```
